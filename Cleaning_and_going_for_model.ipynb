```python
# Identify rows where  less than 0
negative_sales_mask = df['Tax per unit'] < 0

# Change negative  values to their absolute values
df.loc[negative_sales_mask, 'Tax per unit'] = df.loc[negative_sales_mask, 'Tax per unit'].abs()
```

```python
# Identify rows where less than 0
negative_sales_mask = df['Sale Quantity'] < 0

# Change negative  values to their absolute values
df.loc[negative_sales_mask, 'Sale Quantity'] = df.loc[negative_sales_mask, 'Sale Quantity'].abs()
```

```python
# Identify rows where  less than 0
negative_sales_mask = df['TotalQTY '] < 0

# Change negative  values to their absolute values
df.loc[negative_sales_mask, 'TotalQTY '] = df.loc[negative_sales_mask, 'TotalQTY '].abs()
```

```python
# Identify rows where  less than 0
negative_sales_mask = df['TotalSales'] < 0

# Change negative  values to their absolute values
df.loc[negative_sales_mask, 'TotalSales'] = df.loc[negative_sales_mask, 'TotalSales'].abs()
```

```python
df[df['Category ID'] == 271]
```

```python
mean_values_by_category = df.groupby('Category ID')['Gross Value for single Product (Single Sale)'].mean()

# Replace null values with mean values for each category
df['Gross Value for single Product (Single Sale)'] = df.apply(
    lambda row: mean_values_by_category[row['Category ID']]
               if pd.isnull(row['Gross Value for single Product (Single Sale)'])
               else row['Gross Value for single Product (Single Sale)'],
    axis=1
)
```

```python
df.isna().sum()
```

```python
mean_values_by_category = df.groupby('Category ID')['Tax per unit'].mean()

# Replace null values with mean values for each category
df['Tax per unit'] = df.apply(
    lambda row: mean_values_by_category[row['Category ID']]
               if pd.isnull(row['Tax per unit'])
               else row['Tax per unit'],
    axis=1
)
```

```python
mean_values_by_category = df.groupby('Category ID')['Retail Value (Single Sale)'].mean()

# Replace null values with mean values for each category
df['Retail Value (Single Sale)'] = df.apply(
    lambda row: mean_values_by_category[row['Category ID']]
               if pd.isnull(row['Retail Value (Single Sale)'])
               else row['Retail Value (Single Sale)'],
    axis=1
)
```

```python
mean_values_by_category = df.groupby('Category ID')['Buying price per product'].mean()

# Replace null values with mean values for each category
df['Buying price per product'] = df.apply(
    lambda row: mean_values_by_category[row['Category ID']]
               if pd.isnull(row['Buying price per product'])
               else row['Buying price per product'],
    axis=1
)
```

```python
df[df['TotalQTY '].isnull()]
```

```python
#fill TotalQTY null value with Sale Quantity
df['TotalQTY '] = df['TotalQTY '].fillna(df['Sale Quantity'])
```

```python
# cost = Buying price per product * TotalQTY
df['Cost'] = df['Buying price per product'] * df['TotalQTY ']
```

```python
# Tax with Revenue = Tax per unit * Sale Quantity
df['Tax with Revenue'] = df['Tax per unit'] * df['Sale Quantity']
```

```python
#TotalSales = TotalQTY*Gross Value for single Product (Single Sale)
df['TotalSales'] = df['TotalQTY '] * df['Gross Value for single Product (Single Sale)']
```

```python
df.dropna(subset=['Retail Value (Single Sale)'], inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

# Assuming 'df' is your DataFrame, you can select only numeric columns for correlation
numeric_columns = df.select_dtypes(include='number')
correlation_matrix = numeric_columns.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Assuming df is your DataFrame
# df = pd.read_csv('General Sales Data.csv', encoding='utf-8')

# Select relevant columns for correlation matrix (excluding 'Department ID', 'Category ID', and 'Store ID')
columns_to_exclude = ['Department ID', 'Category ID', 'Store ID']
filtered_df = df.drop(columns=columns_to_exclude)

# Compute the correlation matrix for the filtered DataFrame
correlation_matrix = filtered_df.corr(numeric_only=True)

# Set up the figure size
plt.figure(figsize=(10, 8))

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Draw the heatmap with the mask
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', mask=mask)

plt.title('Correlation Matrix with High Correlation and without Department, Category, and Store ID')
plt.show()
```

```python
# Set the size of the figure (width=12, height=6)
plt.figure(figsize=(12, 6))

# Create a histogram using seaborn's histplot
# 'df['Sale Quantity']' is the column you're visualizing
# 'bins=20' specifies the number of bins (bars) in the histogram
# 'kde=True' adds a Kernel Density Estimate (smoothed curve) to the histogram
sns.histplot(df['Sale Quantity'], bins=20, kde=True)

# Set the title of the plot
plt.title('Distribution of Sale Quantity')

# Label the x-axis
plt.xlabel('Sale Quantity')

# Label the y-axis
plt.ylabel('Frequency')

# Display the plot
plt.show()
```
